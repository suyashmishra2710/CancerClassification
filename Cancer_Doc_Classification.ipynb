```python
notes = pd.read_csv('./diagnoses.csv', encoding='latin-1')
notes.head()
```

```python
notes.reset_index()
notes = notes.iloc[:, [1,2]]
notes.columns = ['cancer_type', 'description']
notes.head()
```

```python
#training and testing split
X_train, X_test, y_train, y_test = train_test_split(notes['description'], notes['cancer_type'], stratify=notes['cancer_type'], test_size = 0.25, random_state=42)
```

```python
# transforming the data into numbers using encodings
vectorizer = CountVectorizer(min_df = 0, lowercase= True)
vectorizer.fit(X_train)
X_train = vectorizer.transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
from sklearn.decomposition import TruncatedSVD
model = TruncatedSVD(n_components=5)
X_train = model.fit_transform(X_train)
X_test = model.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
model = StandardScaler(with_mean = False)
X_train = model.fit_transform(X_train)
X_test = model.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score, train_test_split, StratifiedKFold
cv = StratifiedKFold(n_splits = 5, shuffle=True, random_state=42)
model = RandomForestClassifier(n_estimators = 3, max_depth = 10, random_state = 0)




cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
for i, score in enumerate(cv_scores):
    print(f"Fold {i+1} Accuracy: {score:.4f}")

# Print the average accuracy across all folds
print(f"Mean Accuracy: {np.mean(cv_scores):.4f}")

'''
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(model.score(X_test, y_test))
print(classification_report(y_test, y_pred))
'''
```
